```python
pgn = open('lichess_db_standard_rated_2013-01.pgn',encoding="utf-8")
```

```python
dict_of_games = {}
counter = 0
while counter < 1000:
    game = chess.pgn.read_game(pgn)
    dict_of_games[str(counter)] = game
    counter += 1
    if game is None:
        break
```

```python
counter = 0
while counter < 10:
    game = chess.pgn.read_game(pgn)
    if game is None:
        break
    #do stuff here
    for move in game.mainline_moves():
        print(move.uci())
    counter += 1
    print('end of game',counter)
```
